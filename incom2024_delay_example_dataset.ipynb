# 4) Data Analysis
Perform analysis based on our dataset.

```python
print(data.columns)
```

```python
import matplotlib.pyplot as plt

# Example using 'order_item_total_amount'
data['order_item_total_amount'].hist(bins=10)
plt.title('Histogram of Order Item Total Amount')
plt.xlabel('Total Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Line plot: Plot 'order_item_total_amount' on the X-axis and 'profit_per_order' on the Y-axis
plt.plot(data['order_item_total_amount'], data['profit_per_order'])
plt.xlabel('Order Item Total Amount')  # X-axis label
plt.ylabel('Profit Per Order')         # Y-axis label
plt.title('Line Plot of Order Item Total Amount vs Profit Per Order')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Scatter plot: Plot 'sales_per_customer' on the X-axis and 'profit_per_order' on the Y-axis
plt.scatter(data['sales_per_customer'], data['profit_per_order'])
plt.xlabel('Sales Per Customer')  # X-axis label
plt.ylabel('Profit Per Order')    # Y-axis label
plt.title('Scatter Plot of Sales Per Customer vs Profit Per Order')
plt.show()
```

```python
import seaborn as sns

# Box plot: Plot 'order_item_total_amount' to show distribution across 'category_name'
sns.boxplot(x='category_name', y='order_item_total_amount', data=data)
plt.xlabel('Category Name')           # X-axis label
plt.ylabel('Order Item Total Amount') # Y-axis label
plt.title('Box Plot of Order Item Total Amount by Category')
plt.xticks(rotation=45)               # Rotate X-axis labels if needed
plt.show()
```

```python
# Pie chart: Show distribution of 'payment_type'
data['payment_type'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart of Payment Type Distribution')
plt.ylabel('')  # No need for Y-axis label in pie charts
plt.show()
```

```python
import matplotlib.pyplot as plt

# Select the first 5 records from a column (e.g., 'order_item_total_amount')
subset_data = data['order_item_total_amount'].head(5)

# Plotting only the first 5 records
subset_data.plot(kind='bar')
plt.title('First 5 Order Item Total Amounts')
plt.xlabel('Index')
plt.ylabel('Total Amount')
plt.show()
```

```python
# Select the first 10 records using iloc
subset_data = data['order_item_total_amount'].iloc[:10]

# Plot the first 10 records
subset_data.plot(kind='line')
plt.title('First 10 Order Item Total Amounts')
plt.xlabel('Index')
plt.ylabel('Total Amount')
plt.show()
```

```python
# Select 5 random records from a column (e.g., 'order_item_total_amount')
subset_data = data['order_item_total_amount'].sample(5)

# Plot the randomly selected 5 records
subset_data.plot(kind='box')
plt.title('Random 5 Order Item Total Amounts')
plt.xlabel('Random Selection')
plt.ylabel('Total Amount')
plt.show()
```

```python
# Select 10 records from multiple columns
subset_data = data[['order_item_total_amount', 'profit_per_order']].head(10)

# Plotting 'order_item_total_amount' vs 'profit_per_order'
subset_data.plot(kind='scatter', x='order_item_total_amount', y='profit_per_order')
plt.title('Order Item Total Amount vs Profit (First 10 Records)')
plt.xlabel('Order Item Total Amount')
plt.ylabel('Profit Per Order')
plt.show()
```

```python
# Filter the data where 'order_item_total_amount' is greater than 100
subset_data = data[data['order_item_total_amount'] > 100].head(5)

# Plot filtered data
subset_data['order_item_total_amount'].plot(kind='hist')
plt.title('Filtered Order Item Total Amounts > 100 (First 5 Records)')
plt.xlabel('Total Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
# Select the last 5 records using negative indexing
subset_data = data.iloc[-5:]

# Plotting the last 5 records from 'order_item_total_amount'
subset_data['order_item_total_amount'].plot(kind='bar')
plt.title('Last 5 Order Item Total Amounts')
plt.xlabel('Index')
plt.ylabel('Total Amount')
plt.show()
```
